```python
import easyocr
import pandas as pd
import cv2
```

```python
reader = easyocr.Reader(["en"], gpu = True)
```

```python
image = cv2.imread("army_size.png")
result = reader.readtext(image)
print(result)
```

```python
pd.DataFrame(result)
```

```python
for detection in result:
    box = detection[0]
    text = detection[1]
    score = detection[2]

    # Extract the coordinates
    x_min, y_min = map(int, [box[0][0], box[0][1]])
    x_max, y_max = map(int, [box[2][0], box[2][1]])
    
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (50, 255, 0), 2)
    
    # Put the recognized text above the bounding box
    cv2.putText(image, text, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (50, 255, 0), 2)
    print(text)
    
# Display the image with bounding boxes and text
cv2.imshow("OCR Result", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
